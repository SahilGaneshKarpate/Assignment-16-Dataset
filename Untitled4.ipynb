```python
from google.colab.files import upload
import pandas as pd

from google.colab import files
uploaded = files.upload()

df = pd.read_csv("players.csv")

df.head(10)
```

```python
print("Shape:", df.shape)

print("\nDataset Information:")
df.info()

print("\nStatistical Summary:")
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
num_cols = df.select_dtypes(include=["int64","float64"]).columns

cat_cols = df.select_dtypes(include=["object"]).columns

print("Numerical Columns:")
print(num_cols)

print("\nCategorical Columns:")
print(cat_cols)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ["age", "games", "goals", "assists"]

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
cat_cols = ["team", "position", "team_country"]

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(x=df[col])
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=90)
    plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(df.corr(numeric_only=True),
            annot=False,
            cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Average Goals:", df["goals"].mean())

print("Maximum Goals:", df["goals"].max())

print("Minimum Goals:", df["goals"].min())

print("\nAverage Goals by Position:")
print(df.groupby("position")["goals"].mean())
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="position", y="goals", data=df)

plt.title("Goals by Position")
plt.xticks(rotation=45)
plt.show()
```

```python
print("1. Average Age:", round(df["age_years"].mean(),2))

print("2. Average Goals:", round(df["goals"].mean(),2))

print("3. Team with Highest Players:")
print(df["team"].value_counts().idxmax())

print("\n4. Interesting Observations:")
print("- Most players are in their mid-20s.")
print("- Forwards generally score more goals.")
print("- Different teams have different player counts.")
print("- Player performance varies across positions.")
```
